```python
def plot_faces(data,h,w,n_row =3,n_col = 4):
    plt.figure(figsize=(1.8*n_col,2.4*n_row))
    for i in range(n_row * n_col):
        plt.subplot(n_row,n_col,i+1)
        plt.xticks(())
        plt.yticks(())
        plt.imshow(data[i].reshape((h,w)) , cmap = plt.cm.gray)
    plt.show()
```

```python
plot_faces(x, h, w)
```

### 数据集处理

```python
lfw_people = fetch_lfw_people(min_faces_per_person = 70, resize = 0.4)
```

```python
x = lfw_people.data
y = lfw_people.target
```

```python
x.shape,y.shape
```

```python
n_samples, h ,w = lfw_people.images.shape
n_samples, h, w
```

```python
lfw_people.target_names
```

```python
lfw_people.target_names.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.75,random_state = 86)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape,x_test.shape
```

### PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 150, random_state = 86)
```

```python
pca.fit(x_train)
```

```python
pca.components_.shape
```

```python
plot_faces(pca.components_,h,w)
```

```python
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
```

### 人脸识别

```python
from sklearn.svm import SVC
```

```python
clf = SVC(C = 10)
# 参数C = 10: 用于控制模型对训练误差的容忍度（正则化强度的倒数，值越小，正则化越强）
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
clf = SVC(C = 10)
# 参数C = 10: 用于控制模型对训练误差的容忍度（正则化强度的倒数，值越小，正则化越强）
clf.fit(x_train_pca,y_train)
clf.score(x_test_pca,y_test)
```
