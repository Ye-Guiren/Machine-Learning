### 数据集

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w,b = 1.8,2.5
```

```python
np.random.seed(86)
x1 = np.random.rand(100)*4
noise = np.random.randn(100)
x2 = w*x1 +b +noise
```

```python
x1.shape,x2.shape
```

```python
x = np.vstack([x1,x2]).T
x.shape
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
np.mean(x,axis = 0)
```

```python
x -= np.mean(x,axis = 0)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
np.mean(x,axis = 0)
```

### PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 1)
# 将原始数据转换到一维空间，只保留一个主成分（方差最大的方向）
pca.fit(x)
```

```python
pca.components_
# 输出拟合得到的主成分系数
# 表示主成分的方向，是一个二维向量，反映了原始特征在主成分上的权重
```

```python
plt.scatter(x[:,0],x[:,1],s = 10)
plt.plot(
    np.array([pca.components_[0][0]*-1,pca.components_[0][0]])*5,
    np.array([pca.components_[0][1]*-1,pca.components_[0][1]])*5,
    c = 'r'
)
plt.show()
```

```python
x_pca = pca.transform(x)
```

```python
x_pca.shape
```

```python
plt.scatter(x_pca,np.zeros_like(x_pca),s = 10)
plt.show()
```

```python
x_pca_inv = pca.inverse_transform(x_pca)
```

```python
plt.scatter(x[:,0],x[:,1],s = 10)
plt.scatter(x_pca_inv[:,0],x_pca_inv[:,1],s = 10,c = 'r')
plt.show()
```
