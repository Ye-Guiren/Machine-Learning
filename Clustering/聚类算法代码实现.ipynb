```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
X , y= make_blobs(n_samples = 250,centers = 5,n_features=2,random_state=86)
plt.scatter(X[:,0],X[:,1],c = y)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

### K-Means 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5,random_state = 86).fit(X)
```

```python
kmeans.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c = kmeans.labels_)
plt.show()
```

```python
center = kmeans.cluster_centers_
center
```

```python
plt.scatter(X[:,0],X[:,1],c = kmeans.labels_)
plt.scatter(center[:,0],center[:,1],marker = 'x',c = 'red')
plt.show()
```

```python
for n_clusters in [2,3,4,5,6,7]:
    clusterer = KMeans(n_clusters = n_clusters,random_state = 86).fit(X)
    z = clusterer.labels_
    center = clusterer.cluster_centers_
    plt.scatter(X[:,0],X[:,1],c = z)
    plt.scatter(center[:,0],center[:,1],marker='x',c = 'red')
    plt.title("k:{0}".format(n_clusters))
    plt.show()
```

### 

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(linkage = 'ward',n_clusters = 5).fit(X)
# ward()
```

```python
plt.scatter(X[:,0],X[:,1],c=agg.labels_)
# 
plt.show()
```

```python
agg = AgglomerativeClustering(distance_threshold=10,n_clusters=None).fit(X)
# distance_threshold = 10:
# n_clusters=None:distance_threshold
plt.scatter(X[:,0],X[:,1],c = agg.labels_)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
# linkage:
# dendrogram:
def show_dendrogram(model):
    counts = np.zeros(model.children_.shape[0])
    # countsmodel.children_.shape[0]
    n_samples = len(model.labels_)
    for i ,merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
                # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count
    linkage_matrix = np.column_stack(
        [model.children_,model.distances_,counts]
    ).astype(float)
    dendrogram(linkage_matrix)
```

```python
show_dendrogram(agg)
```
