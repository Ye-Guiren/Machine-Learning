## 加载数据集

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
df = pd.read_csv('D:\\LLQ downloads\\ml-master (1)\\ml-master\\datasets\\housing\\housing.csv')
```

```python
california = df

x = california['median_income']
y = california['median_house_value']

x = x[y<500000]
y = y[y<500000]

plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 666)

plt.scatter(x_train,y_train,s=5)
plt.show()
```

## sklearn实现一元回归

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()

x_train = x_train.values 
x_test = x_test.values 
model = lin_reg.fit(x_train.reshape(-1,1),y_train)
```

```python
y_predict = model.predict(x_test.reshape(-1,1))
```

```python
plt.scatter(x_test,y_test,s=10)
plt.plot(x_test,y_predict,c='r')
plt.show()
```

# 模型评价

## 均方误差MSE（Mean Square Error）

```python
y_real = y_test
```

```python
mse = np.sum((y_real - y_predict)**2) / len(y_test)
```

```python
mse
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_real,y_predict)
```

## 均方根误差RMSE（Root Mean Square Error）

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

```python
mean_squared_error(y_real,y_predict,squared = False)
```

## 平均绝对误差MAE（Mean Absolute Error）

```python
mae = np.sum(np.abs(y_real - y_predict)) / len(y_test)
```

```python
mae
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_real,y_predict)
```

## $R^2$ 判定系数 coefficient of determination

```python
r2 = 1 - (np.sum((y_real - y_predict)**2))/ (np.sum((y_real - np.mean(y_real))**2))
```

```python
r2
```

```python
1 - mse/np.var(y_real)
# np.var():求方差
```

```python
from sklearn.metrics import r2_score
r2_score(y_real,y_predict)
```
