### SVR(support vector regression)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(86)
x = np.random.uniform(-3,3,size = 100)
# 生成-3到3之间的100个均匀分布的数
y = 0.5 * x**2+x+2+np.random.normal(0,1,size = 100)
# 添加均值为0、标准差为1的正态分布噪声
X = x.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.svm import LinearSVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
def StandardLinearSVR(epsilon = 0.1):
    # epsilon = 0.1:表示不敏感损失函数的阈值
    # 即允许预测值与真实值之间存在epsilon范围内的误差 
    return Pipeline([
        ("sed_scaler",StandardScaler()),
        ("linearSVR",LinearSVR(epsilon = epsilon))
    ])
```

```python
svr = StandardLinearSVR()
svr.fit(X,y)
y_predict = svr.predict(X)
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color = 'r')
# 对x进行排序，同时按照x的顺序排列预测值y_predict,绘制红色曲线确保曲线按x顺序连贯展示
plt.show()
```

```python
svr.score(X,y)
```

```python
from sklearn.svm import SVR
```

```python
def StandardSVR(epsilon = 0.1):
    return Pipeline([
        ("sed_scaler",StandardScaler()),
        ("SVR",SVR(kernel='rbf',epsilon = epsilon))
    ])
```

```python
svr = StandardSVR()
svr.fit(X,y)
y_predict = svr.predict(X)
plt.scatter(x,y)
plt.plot(np.sort(x),y_predict[np.argsort(x)],color = 'r')
plt.show()
```

```python
svr.score(X,y)
```
