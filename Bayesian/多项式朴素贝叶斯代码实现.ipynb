```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
y = iris.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.8,random_state = 86)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train,y_train)
nb.score(X_test,y_test)
```

```python
X = news.data
y = news.target
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.8,random_state = 86)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# 用于将文本转换成词频向量（即统计每个单词在文本中出现的次数）
vec = CountVectorizer(stop_words = 'english')
# 表示在向量化过程中，自动过滤英文中的停用词，如the、and、is等无实际语义或语义贡献小的词
# 目的是减少特征维度，提升模型训练效率和效果 
```

```python
vec.fit_transform(X)
# fit:学习文本中的词汇表（统计出所有出现的单词）
# transform：将文本转换为词频向量（每个样本表示为单词出现次数的向量）
X_train_vec = vec.transform(X_train)
# 使用已拟合好的vec,将训练集文本X_train转换成词频向量
X_test_vec = vec.transform(X_test)
```
